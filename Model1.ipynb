```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("dataset.csv")

# Data preprocessing and target creation
BASE_PRICE = 10.0

# Calculate occupancy ratio
df['occupancy_ratio'] = df['Occupancy'] / df['Capacity']

# Parameters for pricing formula
alpha = 8.0
queue_weight = 0.5

# Create target variable using linear pricing formula
df['target_price'] = (
    BASE_PRICE + alpha * df['occupancy_ratio'] + queue_weight * df['QueueLength']
)

# Apply the same formula to compute predicted prices
df['predicted_price'] = (
    BASE_PRICE + alpha * df['occupancy_ratio'] + queue_weight * df['QueueLength']
)

# Evaluate using Mean Absolute Error
mae = np.mean(np.abs(df['predicted_price'] - df['target_price']))

# Output results
print(f"Linear Formula Applied | MAE: {mae:.4f} | Avg Price: ${df['predicted_price'].mean():.2f}")

# Simple visualization of Linear Formula Results
plt.figure(figsize=(10, 4))

# Plot 1: Target vs Predicted Price
plt.subplot(1, 2, 1)
plt.scatter(df['target_price'], df['predicted_price'], alpha=0.6, color='blue')
plt.plot(
    [df['target_price'].min(), df['target_price'].max()],
    [df['target_price'].min(), df['target_price'].max()],
    'r--', lw=2
)
plt.xlabel('Target Price')
plt.ylabel('Predicted Price')
plt.title('Target vs Linear Formula Predicted Price')
plt.grid(True, alpha=0.3)

# Plot 2: Price vs Occupancy Ratio
plt.subplot(1, 2, 2)
plt.scatter(
    df['occupancy_ratio'], df['predicted_price'],
    alpha=0.6, label='Linear Formula', color='blue'
)
plt.scatter(
    df['occupancy_ratio'], df['target_price'],
    alpha=0.4, label='Target', color='orange'
)
plt.xlabel('Occupancy Ratio')
plt.ylabel('Price')
plt.title('Price vs Occupancy Ratio')
plt.legend()
plt.grid(True, alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()

```
